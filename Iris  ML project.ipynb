```python
from sklearn.model_selection import train_test_split
x=df.drop(columns=['species'])
y=df['species']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg= LogisticRegression()
```

```python
log_reg.fit(x_train, y_train)
```

```python
#predict
y_pred_train=log_reg.predict(x_train)
y_pred_test=log_reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,roc_curve,auc, roc_auc_score, classification_report, confusion_matrix
accuracy_train=accuracy_score(y_train,y_pred_train)
accuracy_test=accuracy_score(y_test,y_pred_test)
print(accuracy_train)
print()
print(accuracy_test)
```

```python
from sklearn.model_selection import cross_val_score
train_acc=cross_val_score(log_reg, x_train,y_train,cv=10)
test_acc=cross_val_score(log_reg,x_test,y_test,cv=10)
print("Train Accuracy:",train_acc.mean())
print('----'*10)
print("Test Accuracy:",test_acc.mean())
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
```

```python
knn.fit(x_train, y_train)
```

```python
y_pred_train=knn.predict(x_train)
y_pred_test=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,roc_curve,auc, roc_auc_score, classification_report, confusion_matrix
accuracy_train=accuracy_score(y_train,y_pred_train)
accuracy_test=accuracy_score(y_test,y_pred_test)
print("Train:",accuracy_train)
print()
print("Test:",accuracy_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train)
```

```python
y_pred_train=dt.predict(x_train)
y_pred_test=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,roc_curve,auc, roc_auc_score, classification_report, confusion_matrix
accuracy_train=accuracy_score(y_train,y_pred_train)
accuracy_test=accuracy_score(y_test,y_pred_test)
print("Train:",accuracy_train)
print()
print("Test:",accuracy_test)
```
